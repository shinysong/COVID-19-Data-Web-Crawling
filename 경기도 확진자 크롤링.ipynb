```python
from selenium import webdriver 
from urllib.request import urlopen
from bs4 import BeautifulSoup
import pandas as pd
from selenium.webdriver.common.keys import Keys
driver = webdriver.Chrome('C:/chromedriver.exe')
driver.get('https://www.gg.go.kr/bbs/board.do?bsIdx=722&menuId=2903#page=1')
import time
```

```python
a = []
b = []
c = []
d = []
e = []
f = []
g =[]

for n in range(1,100): ##여기 30은 그 다음페이지 넘어가는 거라서 40개씩 보는걸로 해놨기 때문에 만약 검사결과가    
# get titles, address(url)

    res = driver.page_source
    soup = BeautifulSoup(res, "html.parser")
    contents = soup.select('#boardList > tbody > tr')
    for i in range(0,len(contents)):
        content = contents[i]
        a.append(content.select_one('td:nth-child(1)').text)
        b.append(content.select_one('td:nth-child(2)').text)
        c.append(content.select_one('td:nth-child(3)').text)
        d.append(content.select_one('td:nth-child(4)').text)
        e.append(content.select_one('td:nth-child(5)').text)
        f.append(content.select_one('td:nth-child(6)').text)
        g.append(content.select_one('td:nth-child(7)').text)
            
    page_bar = driver.find_elements_by_css_selector('ul.pagination > li > a')
    try:
        if n<=10:
            if n%10 != 0:
                page_bar[n%10+1].click()
                time.sleep(1)
            else:
                page_bar[11].click()
                time.sleep(2)
        else:
            if n%10 != 0:
                page_bar[n%10+2].click()
                time.sleep(1)
            else:
                page_bar[12].click()
                time.sleep(2)
    except:
        print("수집완료")
        break
    time.sleep(0.5)
    
    print(n)
```

```python
data = {'연번':a, '전국번호':b, '확진일자': c, '지역(선별)' : d,
       '발생경위': e,'관련성':f, '격리병원':g}
df = pd.DataFrame(data)
df.describe()
```
